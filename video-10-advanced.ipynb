## 1. Orden de Ejecución de Nodos (Node Execution Order)

Entender la secuencia de ejecución es vital para crear *workflows* donde las ramas dependen unas de otras.

* **Regla de Ejecución (v1.0+):** n8n ejecuta una rama **completamente** antes de pasar a la siguiente [[00:46](http://www.youtube.com/watch?v=TFTLMQLozCI&t=46)].
* **Prioridad:** El orden se define por la posición en el *canvas* (lienzo):
    1.  **De arriba hacia abajo** (*Topmost* a *Bottommost*) [[00:54](http://www.youtube.com/watch?v=TFTLMQLozCI&t=54)].
    2.  Si dos ramas están a la misma altura, se ejecuta primero la **más a la izquierda** (*Leftmost*) [[00:58](http://www.youtube.com/watch?v=TFTLMQLozCI&t=58)].
    * ***Ejemplo Visual:*** El orden es `Wait` (1), `Wait 1` (2), `Wait 2` (3), `Wait 3` (4), seguido de `Wait 4` (5) y `Wait 5` (6), respetando el flujo de arriba a abajo, izquierda a derecha [[01:05](http://www.youtube.com/watch?v=TFTLMQLozCI&t=65)].
* **Recomendación:** Si una rama depende del resultado de otra, es mejor **unir las ramas con un nodo `Merge`** en lugar de cambiar la configuración de la orden de ejecución por defecto [[01:41](http://www.youtube.com/watch?v=TFTLMQLozCI&t=101)].

## 2. División y Unión de Ramas (Splitting and Merging)

### División de Datos

Además del nodo `IF` (que divide en `True`/`False`):

* **Nodo Switch:** Permite crear **N ramas** distintas más una rama de reserva (*fallback*), dividiendo los *Items* de entrada entre ellas según el valor de una clave [[02:21](http://www.youtube.com/watch?v=TFTLMQLozCI&t=141)].
* **Duplicación:** Arrastrar dos o más líneas de salida desde un solo nodo no condicional **duplicará todos los *Items*** a través de cada rama, ejecutando conjuntos de acciones distintas simultáneamente [[02:49](http://www.youtube.com/watch?v=TFTLMQLozCI&t=169)].

### Unión de Datos (Nodo Merge)

El nodo `Merge` es la herramienta clave para combinar flujos, y a diferencia de otros, tiene **dos entradas** [[03:38](http://www.youtube.com/watch?v=TFTLMQLozCI&t=218)]. Espera a que los datos lleguen de ambas entradas antes de continuar.

| Opción de Merge | Descripción | Analogía SQL |
| :--- | :--- | :--- |
| **Append (Añadir)** | Simplemente une los *Items* de ambas entradas, añadiendo los de la segunda a la lista de la primera [[03:50](http://www.youtube.com/watch?v=TFTLMQLozCI&t=230)]. | **UNION** |
| **Choose (Elegir)** | Permite seleccionar qué entrada pasar (Input 1, Input 2, o ninguna). Útil para **sincronizar** y asegurar que ambas ramas terminen antes de continuar el *workflow* [[04:01](http://www.youtube.com/watch?v=TFTLMQLozCI&t=241)]. | N/A |
| **Combine (Combinar)** | Une los *Items* de ambas entradas basándose en una **clave de coincidencia** (por posición o por campos) [[04:15](http://www.youtube.com/watch?v=TFTLMQLozCI&t=255)]. | **JOIN** (Inner, Outer, Left, Right) [[04:31](http://www.youtube.com/watch?v=TFTLMQLozCI&t=271)]. |

#### 🔎 Caso de Uso Avanzado: Enriquecimiento de Datos (*Enrichment*)

Se utiliza la opción **`Combine`** para enriquecer un flujo de datos (ej. contactos) con datos de otro flujo (ej. compañías) [[11:17](http://www.youtube.com/watch?v=TFTLMQLozCI&t=677)]:

1.  **Preparación:** En un flujo se extraen los contactos profesionales. En otro flujo se leen los datos de las compañías (sitio web, país).
2.  **Clave de Unión:** Se utiliza el **dominio** de correo (extraído del email) como clave en el flujo de contactos, y la clave **`website`** en el flujo de compañías [[12:01](http://www.youtube.com/watch?v=TFTLMQLozCI&t=721)].
3.  **Merge:** Se configura el nodo `Merge` para **`Combine`** por campos (`domain` de un lado, `website` del otro) y se selecciona **"Keep all of Input 2 with data from Input 1 added in"** [[13:08](http://www.youtube.com/watch?v=TFTLMQLozCI&t=788)].
4.  **Resultado:** Los *Items* de contacto se **enriquecen** con los detalles de la compañía que coincidió con el dominio [[14:11](http://www.youtube.com/watch?v=TFTLMQLozCI&t=851)].

## 3. Iteración sobre Múltiples Items (Loop Over Items)

Aunque n8n ejecuta los nodos una vez por *Item* de forma predeterminada, el nodo `Loop Over Items` (o `Split in Batches`) es necesario para el manejo de conjuntos de datos dentro de un *workflow* [[05:08](http://www.youtube.com/watch?v=TFTLMQLozCI&t=308)].

* **Función:** Divide el flujo de *Items* de entrada en **lotes (*batches*)** de un tamaño específico (frecuentemente tamaño 1), ejecutando un segmento del *workflow* (el *loop*) para cada lote [[05:22](http://www.youtube.com/watch?v=TFTLMQLozCI&t=322)].
* **Casos de Uso Esenciales:**
    * **Manejar Múltiples Hojas/Fuentes:** Si recibes múltiples *Items* (ej. varias cuentas) y cada *Item* debe activar una acción que retorna N resultados (ej. leer una hoja de cálculo completa para cada cuenta) [[06:00](http://www.youtube.com/watch?v=TFTLMQLozCI&t=360)].
    * **Limitar la Tasa de API (*Rate Limits*):** Dividir los *Items* en lotes pequeños y colocar un nodo **`Wait`** (Esperar) dentro del *loop* para evitar sobrecargar un API con demasiadas llamadas simultáneas [[06:38](http://www.youtube.com/watch?v=TFTLMQLozCI&t=398)].
