## 👥 Colaboración y Seguridad en n8n

El video aborda cómo interactuar con la comunidad, cómo compartir el trabajo de forma segura y cómo usar la propia API de n8n para la gestión de *workflows* avanzados.

### 1. La Comunidad y la Biblioteca de Plantillas 💡

* **Comunidad n8n:** Es una plataforma activa donde puedes pedir ayuda, reportar *bugs* (errores), solicitar nuevas funcionalidades (*feature requests*) o encontrar anuncios [[00:17](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=17)].
    * La comunidad es fundamental para obtener respuestas rápidas y prevenir tiempos de inactividad.
* **Plantillas (*Templates*):** La biblioteca de plantillas cubre una amplia gama de problemas de automatización ya resueltos (ej. crear un *endpoint* de API o respaldar *workflows* en GitHub) [[01:21](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=81)].
    * **Mejor Práctica:** **Siempre consulta la biblioteca** antes de empezar a construir un *workflow* desde cero, ya que podría ahorrarte mucho tiempo [[01:49](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=109)].
    * Puedes enviar tus propios *workflows* como plantillas, y n8n garantiza que **no se comparten credenciales sensibles** [[02:44](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=164)].

### 2. Gestión de Usuarios y Roles de Acceso 🔒

n8n utiliza un sistema de roles para controlar quién puede acceder y modificar *workflows* y credenciales.

#### A. Tipos de Usuarios [[03:16](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=196)]

| Rol | Descripción | Acceso a Información Sensible |
| :--- | :--- | :--- |
| **Owner (Propietario)** | Hay uno por instancia. Tiene todos los derechos administrativos. | No puede leer la **información sensible** subyacente (ej. la *API Key* real) dentro de una Credencial [[03:51](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=231)]. |
| **Admin (Administrador)** | Similar al Owner, pero sin acceso al *dashboard* de n8n Cloud o a la gestión del rol Owner. | No puede leer la información sensible subyacente de las credenciales. |
| **Member (Miembro)** | Usuario estándar. Gestiona sus propias cuentas y *workflows*. | Acceso limitado por las configuraciones de compartición. |

#### B. Compartición de Workflows y Credenciales

* **Compartición de Workflows** [[05:35](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=335)]:
    * Permite a otros usuarios de la misma instancia acceder a tus *workflows*.
    * **Roles en el *Workflow*:** **Creator** (el que lo creó) y **Editor** (usuarios con permiso de modificación).
    * **Implicación Crítica:** Al compartir un *workflow*, los Editores obtienen permiso para **utilizar todas las Credenciales** que este *workflow* utiliza, **incluso si no se les compartió la credencial por separado** [[06:50](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=410)]. Esto es para permitirles probar y depurar, pero subraya la importancia de la seguridad.
* **Compartición de Credenciales** [[07:42](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=462)]:
    * Permite que otros usuarios usen la Credencial en *sus propios workflows*.
    * **Seguridad:** El usuario compartido **no puede acceder ni editar los detalles sensibles** de la credencial (la API Key), solo puede usarla [[07:53](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=473)].
    * Se gestiona desde el menú de Credenciales.

### 3. La API de n8n y el Nodo n8n 🤖

n8n tiene su propia API, lo que permite automatizar la plataforma desde dentro o fuera de la herramienta [[08:57](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=537)].

* **Usos Principales:**
    1.  **Gestión Externa:** Acceder, crear, editar o eliminar *workflows* y credenciales desde un programa externo (*Rest API*).
    2.  **Gestión Interna (Nodo n8n):** Usar el nodo `n8n` para que un *workflow* **gestione a otros *workflows*** o a sí mismo de manera programática [[10:18](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=618)].
* **Caso de Uso Programático Avanzado:** Un *workflow* de soporte puede **desactivar** un *workflow* principal a las 6:00 p.m. y **reactivarlo** a las 8:00 a.m. (o hacerlo directamente en el *workflow* principal con lógica condicional) [[10:40](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=640)].
* **Autenticación:** Debes generar una **API Key** en la sección "n8n API" de la Configuración para autenticarte en la API [[09:37](http://www.youtube.com/watch?v=pI0W-0Qcwmo&t=577)].
